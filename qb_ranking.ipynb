```python
import pandas as pd
from scipy.sparse import lil_matrix
import scipy.spatial.distance
```

```python
stats = pd.read_csv('Full_NFL_Stats_2022.csv')
```

```python
stats.head()
```

```python
stats.groupby('teamName')
```

```python
stats.sort_values('rank',ascending=True)
```

```python
stats.dropna(subset='rank',inplace=True)
```

```python
stats.sort_values('rank',ascending=True)
```

```python
stats.drop(columns=['perGameValue', 'Unnamed: 0'],axis=1, inplace=True)
```

```python
rows = stats.loc[stats['rank'] > 32]
print(rows)
```

```python
interceptions = stats.loc[stats['displayName'] == 'Interception Percentage']
interceptions = interceptions.sort_values(by='rank')
```

```python
completion = stats.loc[stats['displayName'] == 'Completion Percentage']
completion = completion.sort_values(by='rank')
```

```python
pass_rating = stats.loc[stats['displayName'] == 'Passer Rating']
pass_rating = pass_rating.sort_values(by='value',ascending=False)
```

```python
touchdowns = stats.loc[stats['abbreviation'] == 'TD%']
touchdowns = touchdowns.drop('rank',axis=1)
touchdowns = touchdowns.sort_values(by='value',ascending=False)
```

```python
pass_yard_per_game = stats.loc[stats['displayName'] == 'Passing Yards Per Game']
pass_yard_per_game = pass_yard_per_game.sort_values(by='rank',ascending=True)
```

```python
bottom_qb = []

for df in [interceptions, completion, pass_rating, touchdowns, pass_yard_per_game]:

    bottom_names = list(df['teamName'].tail(16))

    bottom_qb += bottom_names

name_counts = {name: bottom_qb.count(name) for name in set(bottom_qb)}

final_names = [name for name in name_counts.keys() if name_counts[name] == 5]

print(final_names)
```
