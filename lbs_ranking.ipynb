```python
stats.drop(columns=['perGameValue', 'Unnamed: 0'],axis=1, inplace=True)
```

```python
rows = stats.loc[stats['rank'] > 32]
print(rows)
```

```python
sacks = stats.loc[stats['displayName'] == 'Sacks']
sacks = sacks.sort_values(by='rank')
sacks
```

```python
tottackles = stats.loc[stats['displayName'] == 'Total Tackles']
tottackles = tottackles.sort_values(by='value', ascending=False)
tottackles
```

```python
interceptions = stats.loc[stats['description'] == 'The total number of interceptions.']
interceptions = interceptions.sort_values(by='rank')
interceptions
```

```python
tfl = stats.loc[stats['displayName'] == 'Tackles For Loss']
tfl = tfl.sort_values(by='rank')
tfl
```

```python
bottom_lb = []

for df in [interceptions, tottackles, tfl, sacks]:

    bottom_names = list(df['teamName'].tail(16))

    bottom_lb += bottom_names

name_counts = {name: bottom_lb.count(name) for name in set(bottom_lb)}

final_names = [name for name in name_counts.keys() if name_counts[name] == 4]

print(final_names)
```
